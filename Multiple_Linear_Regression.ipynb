```python
#import libraries what i want this project

import pandas as pd;
import numpy as np;
import matplotlib.pyplot as mat;

```

```python
# Read "csv" file and returns in table format.

clientData = pd.read_csv("insurance_pre.csv");

clientData
```

```python
# "get_dummies" - convert catagorical datas to numbers

clientData = pd.get_dummies(clientData, drop_first = True)

clientData
```

```python
clientData.columns
```

```python
independentValue = clientData[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']];

# Independent catagorical columns(sex, smoker) are nominal data(we cannot able to compare) so go with one hot encoding. 
independentValue  
```

```python

dependentValue = clientData[["charges"]]

dependentValue
```

```python
#mat.scatter(independentValue, dependentValue);
```

```python

#Split Train and Test

from sklearn.model_selection import train_test_split;

x_train, x_test, y_train, y_test = train_test_split(independentValue, dependentValue, test_size = 1/3, random_state = 0)

x_train

```

```python
y_train

```

```python
x_test
```

```python
y_test
```

```python
# Model Creation


from sklearn.linear_model import LinearRegression;

regressor = LinearRegression(fit_intercept = True);

regressor.fit(x_train, y_train);
```

```python
weight = regressor.coef_



print("Weight of the model = {}".format(weight));

weight
```

```python
bias = regressor.intercept_

print("Bias of the model = {}".format(bias));

bias
```

```python
y_pred = regressor.predict(x_test);

#y_pred
```

```python
#Evaluation metrics

from sklearn.metrics import r2_score;

metricValue = r2_score(y_test, y_pred);

metricValue    
```

```python
#save model

import pickle;

filename = "MultipleLinear_FinalModel.sav";

pickle.dump(regressor, open(filename, "wb"));

```

```python
#load model

loadModel = pickle.load(open(filename, "rb"));


```
