```python

import pandas as pd;
import numpy as np;
import matplotlib.pyplot as mat;

```

```python

clientInformation = pd.read_csv("insurance_pre.csv");

clientInformation
```

```python
# convert catagorical columns to numbers

clientInformation = pd.get_dummies(clientInformation, drop_first = True);

clientInformation
```

```python

independentValue = clientInformation[["age", "bmi", "children", "sex_male", "smoker_yes"]];

independentValue
```

```python
dependentValue = clientInformation[["charges"]];

dependentValue
```

```python

from sklearn.model_selection import train_test_split;

x_train, x_test, y_train, y_test = train_test_split(independentValue, dependentValue, test_size = 1/3, random_state = 0);
```

```python
x_train
```

```python
#import Polynominal library from Documentation

#np.poly1d(np.polyfit(x_train,y_train,3));

#np.linspace(1, 22, 100);


from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LinearRegression;

linearRegressor  = LinearRegression();

linearRegressor.fit(x_train, y_train);

#np.zeros((156816, 36, 53806), dtype='uint8');

#from sklearn import Linear_model
#regressor = PolynomialFeatures();

regressor = PolynomialFeatures(degree = 2);

x_train_transformed = regressor.fit_transform(x_train);

x_test_transformed = regressor.fit_transform(x_test);

#x_test = regressor.fit_transform(x_test);

regressor.fit(x_train_transformed, y_train);

linear_2 = LinearRegression();

linear_2.fit(x_train_transformed, y_train);

```

```python

y_pred = linear_2.predict(x_test_transformed);

#y_pred
```

```python

from sklearn.metrics import r2_score;

metricscore = r2_score(y_test, y_pred);

metricscore
```
